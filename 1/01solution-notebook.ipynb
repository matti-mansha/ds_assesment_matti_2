```python
df_work = df.copy()
```

```python
df_work.drop(columns=['ADDRESSLINE2', 'STATE', 'TERRITORY'], inplace=True)
```

```python
df_work.dropna(axis=0, inplace=True)
```

```python
columns = df_work.columns
df_work.columns
```

```python
# convert the date column to a datetime data type
df_work['ORDERDATE'] = pd.to_datetime(df_work['ORDERDATE'])
df_work['ORDERDATE'].value_counts()
```

```python
# convert the "STATUS" column to a categorical data type
df_work['STATUS'] = pd.Categorical(df_work['STATUS'])
df_work['STATUS'].value_counts()
```

```python
# convert the "STATUS" column to a categorical data type
df_work['PRODUCTLINE'] = pd.Categorical(df_work['PRODUCTLINE'])
df_work['PRODUCTLINE'].value_counts()
```

```python
df_work['PRODUCTCODE'] = pd.Categorical(df_work['PRODUCTCODE'])
df_work['PRODUCTCODE'].value_counts()
```

```python
# define a regular expression to match phone numbers
regex = r'\D+'

# standardize phone numbers by removing all non-numeric characters
df_work['PHONE'] = df_work['PHONE'].replace(regex, '', regex=True)

# insert separators into the standardized phone numbers
df_work['PHONE'] = df_work['PHONE'].str.replace(r'(\d{3})(\d{3})(\d{4})', r'(\1) \2-\3')
```

```python
df_work['PHONE']
```

```python
import re

def standardize_phone_number(phone):
    # remove all non-numeric characters
    phone = re.sub(r'\D+', '', phone)
    
    # insert separators into the standardized phone number
    if len(phone) == 10:
        phone = re.sub(r'(\d{3})(\d{3})(\d{4})', r'(\1) \2-\3', phone)
    elif len(phone) == 7:
        phone = re.sub(r'(\d{3})(\d{4})', r'\1-\2', phone)
    else:
        phone = 'Invalid phone number'
    
    return phone
```

```python
# apply the function to the age column using map()
df_work['PHONE'] = df_work['PHONE'].map(standardize_phone_number)
```

```python
df_work['PHONE']
```

```python
df_work.drop(columns=['PHONE'], inplace=True)
```

```python
df_work['MSRP'] = df_work['MSRP'].astype('int')
df_work['MSRP'].value_counts()
```

```python
df_work['CITY'] = pd.Categorical(df_work['CITY'])
df_work['CITY'].value_counts()
```

```python
df_work['COUNTRY'] = pd.Categorical(df_work['COUNTRY'])
df_work['COUNTRY'].value_counts()
```

```python
df_work.drop(columns=['POSTALCODE'], inplace=True)
```

```python
df_work['DEALSIZE'] = pd.Categorical(df_work['DEALSIZE'])
df_work['DEALSIZE'].value_counts()
```
